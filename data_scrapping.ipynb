```python
data = data.drop(columns = ['events', 'height', 'weight', 'prev_3_medals', 'country'])
data[data.isnull().any(axis=1)]
```

```python
data = data.dropna()
data
```

```python
train = data[data['year'] < 2012].copy()
test = data[data['year'] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
predictors = ['athletes', 'prev_medals']
target = 'medals'
```

```python
reg.fit(train[predictors], train[target])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions
```

```python
test['predictions'] = predictions
```

```python
test
```

```python
test.loc[test['predictions'] < 0, 'predictions'] = 0
test['predictions'] = test['predictions'].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test['medals'], test['predictions'])
```

```python
error
```

```python
data.describe()['medals']
```

```python
errors = (test['predictions'] - test['medals']).abs()
```

```python
team_error = errors.groupby(test['team']).mean()
```

```python
team_error
```

```python
medals = data['medals'].groupby(data['team']).mean()
```

```python
medals
```

```python
error_ratio = team_error / medals
```

```python
error_ratio
```

```python
error_ratio[~pd.isnull(error_ratio)]
```

```python
import numpy as np
error_ratio = error_ratio[np.isfinite(error_ratio)]
```

```python
error_ratio
```

```python
error_ratio.plot.hist()
```

```python
error_ratio.sort_values(ascending=False)
```
